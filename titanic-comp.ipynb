```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder

from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
import os
print(os.listdir('/kaggle/input/titanic-dataset'))
```

```python
df = pd.read_csv('/kaggle/input/titanic-dataset/Titanic-Dataset.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna({'Age': df['Age'].mean(), 'Cabin': 'U'}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Embarked'])
```

```python
df.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Survived'], axis=1),
                                                   df['Survived'],
                                                   test_size=0.2,
                                                   random_state=42)
```

```python
X_train.head()
```

```python
clf = LogisticRegression()

ohe = OneHotEncoder(handle_unknown='ignore')
```

```python
X_train_transformed = ohe.fit_transform(X_train)
X_test_transformed = ohe.transform(X_test)
```

```python
clf.fit(X_train_transformed, y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
```

```python
accuracy_score(y_pred, y_test)
```

```python
X = df.drop('Survived', axis=1) 
y = df['Survived']
X_transformed = ohe.fit_transform(X)
```

```python
np.mean(cross_val_score(clf,X_transformed,y,cv=10))
```

```python
clf.fit(X_transformed, y)  # Train on all data
```

```python
submission = pd.DataFrame({
    'PassengerId': X_test['PassengerId'],
    'Survived': y_pred
})
```

```python
submission.to_csv('submission.csv', index=False)
```
